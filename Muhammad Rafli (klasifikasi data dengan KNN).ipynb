# Klasifikasi data dengan Algoritma K-NN

```python
#memanggil library yang akan digunakan

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#membaca dataset

dataset = pd.read_csv('Social Network Ads.csv')
```

```python
#menampilkan dataset 5 yang paling atas

dataset.head()
```

```python
#membagi dataset kedalam X dan Y 

x = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:, -1].values
```

```python
#menampilkan seluruh dataset yang tebagi dalam X

print(x)
```

```python
#membagi dataset kedalam data training dan data testing menggunakan train_test_split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
#menampilkan data training

print(x_train)
```

```python
#menjumlahkan seluruh data 

len(x)
```

```python
#menjumlahkan data training yang telah di bagi

len(x_train)
```

```python
#menjumlahkan data testing yang telah di bagi

len(x_test)
```

```python
#melakukan scaling data karena jarak data yang terlalu besar

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#menampilkan data training yang telah di scaling

print(x_train)
```

```python
#menampilkan data testing yang telah di scaling

print(x_test)
```

```python
#modelling data dengan menggunakan algoritma KNN

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
classifier.fit(x_train, y_train)
```

```python
#menentukan prediksi dari data x testing

y_pred = classifier.predict(x_test)
```

```python
#hitung evaluasi/validasi menggunakan confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#visualisasi gambar confusion matrix

import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(8,5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt=".0f", ax=ax)
plt.xlabel("y_head")
plt.ylabel("y_true")
plt.show()
```

```python
#cross validation dataset dengan menggunakan recall, precision, f1-score

from sklearn.metrics import f1_score, precision_score, recall_score 

print('Recall: {}'.format(recall_score(y_test, y_pred)))
print('Precision: {}'.format(precision_score(y_test, y_pred)))
print('F1-Score: {}'.format(f1_score(y_test, y_pred)))
```

```python
#cross validation dataset dengan menggunakan recall, precision, f1-score

from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred))
```

```python
#menampilkan visual gambar persebaran data training

from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min()-1, stop=x_set[:, 0].max()+1, step=0.01,),
                     np.arange(start=x_set[:, 1].min()-1, stop=x_set[:, 1].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape), 
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate (np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
            c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan K-NN (data training)')
plt.xlabel('umur')
plt.ylabel('Estimasi Gaji')
plt.legend()
plt.show()
```

```python
#menampilkan visual gambar persebaran data testing

from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min()-1, stop=x_set[:, 0].max()+1, step=0.01,),
                     np.arange(start=x_set[:, 1].min()-1, stop=x_set[:, 1].max()+1, step=0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape), 
             alpha = 0.75, cmap = ListedColormap(('blue', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate (np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
            c = ListedColormap(('blue', 'green'))(i), label = j)
plt.title('Klasifikasi Data dengan K-NN (data testing)')
plt.xlabel('umur')
plt.ylabel('Estimasi Gaji')
plt.legend()
plt.show()
```
